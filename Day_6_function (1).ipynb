```python
def cal(a,b):
  """addition,subtraction,multiplacition,division"""
  choice=input("please enter your choice...1)add ,2)sub, 3)mul, 4)div")
  if choice == "1":
    return a+b
  elif choice=="2":
    return a-b
  elif choice=="3":
    return a*b
  elif choice == "4":
    try:
      return a/b
    except ZeroDivisionError:
      return "cannot divid by zero"
  else:
    return "invalid choice"
```

```python
cal(5,2)
```

Q) By using function write a code for to print a rightangle triangle or dimond shapes by taking a choice from the user

```python
def cal(n):
  """rightangle triangle, dimond"""
  choice=input("please enter your choice...1)right_triangle, 2)dimond")
  if choice == "1":
    for i in range(n)         :
      for j in range(i+1):
        print("*",end=" ")
      print()
  if choice=="2":
     for i in range(1,n+1):
      if i==1:
        print((n-i)*" " +"*")
      else:
        print((n-i)*" "+"*"+" "*(2*i-3)+"*")
     for i in range(n-1,0,-1):
      if i==1:
        print((n-i)*" " +"*")
      else:
        print((n-i)*" "+"*"+" "*(2*i-3)+"*")

```

```python
cal(5)
```
